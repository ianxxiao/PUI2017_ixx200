### 1. Get Environment Variable

```python
import os
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

PUIdata = os.getenv('DFDATA')
print (PUIdata)
```

```python
path = PUIdata + '/h9gi-nx95/1428430582/h9gi-nx95'
print(path)
```

```python
df = pd.read_csv(path)
df.head(10)
```

```python
df.dtypes
```

### 2. Keep Only Date and Count Columns

```python
keepCol =['DATE', 'NUMBER OF PEDESTRIANS INJURED']
ADS = df[keepCol]
ADS.head(10)
```

```python
ADS.columns =['Date', 'Num_Pedestrian_Injured']
```

### 3.Create a Analytics Dataset by Aggregating by Date

```python
sumInjury = ADS.groupby('Date').sum().reset_index()
sumInjury.columns = ['Date', 'Total_Pedestrian_Injury']
sumInjury.head(20)
```

```python
sumInjury['Date'] = pd.to_datetime(sumInjury.Date)
sumInjury.dtypes
```

```python
sumInjury.sort('Date', inplace = True)
sumInjury.head(20)
```

### 4.Plot a Time Series Graph

```python
plt.plot(sumInjury.Date, sumInjury.Total_Pedestrian_Injury)
plt.title("Total Pedestrian Injuries by Days (2012-2015)")
plt.xlabel("Date")
plt.ylabel("Count of Pedestrian Injury")
plt.xticks(rotation = 70)
```
